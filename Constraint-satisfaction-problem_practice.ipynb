### A Python program to graph coloring the attached graph with only 3 colors (Red, Green, and Blue), and adjacent nodes must have different colors using backtracking

#### Algorithm

The steps required to color a graph G with n number of vertices are as follows −
Arrange the vertices of the graph in some order.
Choose the first vertex and color it with the first color.
Choose the next vertex and color it with the lowest numbered color that has
not been colored on any vertices adjacent to it. If all the adjacent vertices are
colored with this color, assign a new color to it. Repeat this step until all the
vertices are colored.
And for printing all the possible solutions after calling the function we will
undo our changes which is the main concept of backtracking everytime so
that it discovers all the possible solutions.
